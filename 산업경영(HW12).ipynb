```python
# 정점이름: start, A, B, C, D, END    //출발점 위 아래 위 아래 도착점

graph = {}
graph['start'] = {}
graph['start']['A'] = 5
graph['start']['B'] = 2

graph['A'] = {}
graph['A']['C'] = 4
graph['A']['D'] = 2

graph['B'] = {}
graph['B']['A'] = 8
graph['B']['D'] = 7

graph['C'] = {}
graph['C']['D'] = 6
graph['C']['end'] = 3

graph['D'] = {}
graph['D']['end'] = 1

graph['end'] = {}

# 초기 비용설정
infinity = float('inf')
costs = {}
costs['A'] = 5
costs['B'] = 2
costs['C'] = infinity
costs['D'] = infinity
costs['end'] = infinity

# 부모노드
parents = {}
parents['start'] = None
parents['A'] = 'start'
parents['B'] = 'start'
parents['C'] = 'A'
parents['D'] = None
parents['end'] = None

# 이동경로
processed = []


def find_lowest_cost_node(costs):
    lowest_cost = float('inf')
    lowest_cost_node = None
    global processed
    
    for node in costs:
        cost = costs[node]
        if cost < lowest_cost and node not in processed:
            lowest_cost = cost
            lowest_cost_node = node
            
    return lowest_cost_node


# 알고리즘
node = find_lowest_cost_node(costs)
while node is not None:
    cost = costs[node]
    neighbors = graph[node]

    for n in neighbors.keys():
        new_cost = cost + neighbors[n]
        if new_cost < costs[n]:
            costs[n] = new_cost
            parents[n] = node
            
    processed.append(node)    # 주의! 
    node = find_lowest_cost_node(costs)
    

# 최단경로 표시
# print(parents)

sp_list = []
current = 'end'
while current != None: # current에 항목이 있으면 진행
    sp_list.append(current)
    current = parents[current]
    
sp_list.reverse()
sp_list = ' -> '.join(sp_list)
print('최단경로: ', sp_list, ', ', 'total: ',costs['end'])
```
